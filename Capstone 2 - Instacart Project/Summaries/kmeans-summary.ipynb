# **Customer Segmentation using KMeans Clustering - Summary**

## Introduction
#### In this summary analysis, we segmented the customers (represented in the data by user_id) into distinct purchasing behavioral groups using KMeans clustering.  Doing this is foundational and pivotal in establishing what marketing promotions and strategies we wish to employ further down the project pipeline.

## Data Preprocessing and Feature Creation
#### Before applying KMeans, we performed essential data preprocessing steps such as scaling the features to ensure equal importance and handling missing values to maintain data integrity. Key features were selected based on their relevance to customer behavior, including order_dow_mean, order_hour_of_day_mean, time_between_purchases, and purchase_frequency.

## Clustering Results
#### We used KMeans clustering with an optimal k of 3 clusters. Here are the key metrics for the clustering performance:

#### Silhouette Coefficient: 0.3521

#### WCSS (Within-Cluster Sum of Squares): 473,425.83

#### BCSS (Between-Cluster Sum of Squares): 351,379.47

#### The metrics by themselves don't mean as much right now other than to present, however once our other clustering methods are completed we will be able to input them into a table and conduct a comparison to select our best clustering method for this project.

## Conclusion
#### Based on the clustering results, we recommend the following actions (pending selection of this clustering method):

#### Cluster 0: Implement loyalty programs to maintain their shown high level of engagement.

#### Cluster 1: Offer time-specific promotions to encourage purchases during off-peak hours.

#### Cluster 2:  Send personalized reminders to reduce the time between purchases.

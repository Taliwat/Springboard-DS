## Introduction
#### In this summary analysis, we segmented the customers once again(represented in the data by user_id) into distinct purchasing behavioral groups using Agglomerative clustering.  We will go through much of the same process as the KMeans clustering one to see how this clustering method compares.

## Data Preprocessing and Feature Creation
#### Also similar to our KMeans analysis we will be using the same preprocessing and features to cluster and segment the data (order_dow_mean, order_hour_of_day_mean, time_between_purchases, and purchase_frequency).

## Clustering Results
#### We used KMeans clustering with an optimal k of 3 clusters. Here are the key metrics for the clustering performance:

#### Silhouette Coefficient: 0.2776

#### WCSS (Within-Cluster Sum of Squares): 28520.669

#### BCSS (Between-Cluster Sum of Squares): 15327.630

#### The metrics by themselves don't mean as much right now other than to present, however once our other clustering methods are completed we will be able to input them into a table and conduct a comparison to select our best clustering method for this project.

#### Based on the above heatmap, we can see customer segmentation and characterization from our 3 clusters. Here is what can be observed:

#### Cluster 0 (Infrequent Shoppers): Customers in this cluster tend to purchase less frequently and order evenly throughout the week.  They tend to have the longest intervals between purchases and the lowest purchase frequency.

#### Cluster 1 (Average/Moderate Shoppers): This cluster represents customers with average or balanced shopping patterns.  They have have moderate intervals between purchases and an average purchasing frequency.

#### Cluster 2 (Frequent Shoppers): These shoppers are the most frequent, with the shortest intervals between purchases and the highest purchase frequency.

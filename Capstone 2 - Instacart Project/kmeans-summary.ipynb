# Customer Segmentation using KMeans Clustering - Summary

## Introduction - In this summary analysis, we segmented the customers (represented in the data by user_id) into distinct purchasing behavioral groups using KMeans clustering.  Doing this is foundational and pivotal in establishing what marketing promotions and strategies we wish to employ further down the project pipeline.

## Data Preprocessing - To ensure the quality of our clustering results, we preprocessed the data.  What this means is we handled any missing values in the data presented, and scaled the values so that our data behaved more efficiently.

## Feature Creation - The features (measures for what we want to base our customer groups on) were created from the data.  What we are measuring the customer groups and extracting them on are based on this criteria:
#### -Their purchase frequency
#### -Avg time between purchases
#### -Avg time of day of user purchases
#### -Avg day of user purchases

## Determining the Optimal Number of Clusters
## We used the elbow method and silhouette score and plot to determine the optimal number of clusters (read customer groups) for this KMeans analysis.

### Elbow Method Results


#### Looks like at least from this elbow plot like the best k value is k=4 due to how the plot starts to 'flatten' at that point.  This means it is looking like the data shows 4 separate customer groups.

### Elbow Plot using Silhouette Scores


Similar to the Elbow plot above, this plot also can offer some validation to our k = 4 value showing again the flattening of the curve at the k = 4 marker.  This plot differs only in that it uses a different metric in the Silhouette score.

### Silhouette Plot

#### In the plot above, you can see 3 very similar clusters in size, with little overlap (meaning customers won't belong to more than one group so the spacing is good).  We would like all the clusters to be pretty uniform in size when observing this plot, and 3 of our clusters do this.

## Individual Cluster Characteristics - We will now provide identity to each cluster that we have shown, in order to give value to them and make it easier to identify with throughout this project.

## Cluster 0 - Frequent Shoppers

#### These users shop regularly and frequently, placing orders mostly early in the week and during mid-day hours.

#### Early scope recommendation:  These customers could be targeted with frequent shopper rewards or loyalty programs.

## Cluster 1 - Night Shoppers

#### These users prefer to shop later in the day, often at night.

#### Early scope recommendation:  Marketing strategies could include promotions that target late-night sales.

## Cluster 2 - Weekend Shoppers

#### These users primarily shop during the weekends, especially in the mornings.

#### Early scope recommendation:  Promotions for weekend deals or early bird specials could prove to be effective.

## Cluster 3 - Occasional Shoppers

#### These users shop less frequently and their orders are spread throughout the week and day.

#### Early scope recommendation:  This group might respond well to occasional discounts or personalized offers in order to increase their shopping frequency.
